```python
import numpy as np
import matplotlib.pyplot as plt

def separating_line(x1, x2):
    return -0.25 * x1 - 0.5 - x2

x1_range = np.linspace(-5, 5, 100)
x2_range = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1_range, x2_range)

f = separating_line(X1, X2)

plt.figure(figsize=(8, 7))

plt.contourf(X1, X2, f, levels=[-10, 0, 10], 
             colors=['#add8e6', '#ffb6c1'], alpha=0.5)

plt.contour(X1, X2, f, levels=[0], colors='black', linewidths=2)

plt.plot([-2], [0], 'ko', markersize=8, label='Point (-2, 0)')
plt.plot([0], [-0.5], 'ro', markersize=8, label='Point (0, -0.5)')

plt.xlabel('$x_1$', fontsize=12)
plt.ylabel('$x_2$', fontsize=12)

plt.text(2, 2, 'sign(w$^T$x) = +1\n(Region where w$^T$x > 0)', 
         horizontalalignment='center', color='black')
plt.text(-2, -2, 'sign(w$^T$x) = -1\n(Region where w$^T$x < 0)', 
         horizontalalignment='center', color='black')

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')

plt.savefig('homework_1_plot.png', dpi=300)

plt.show()
```

```python
def separating_line(x1, x2):
    return 0.5 * x1 + 1 - x2

x1_range = np.linspace(-5, 5, 100)
x2_range = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1_range, x2_range)

f = separating_line(X1, X2)

plt.figure(figsize=(8, 7))

plt.contourf(X1, X2, f, levels=[-10, 0, 10], 
             colors=['#ffb6c1', '#add8e6'], alpha=0.5)

plt.contour(X1, X2, f, levels=[0], colors='black', linewidths=2)

plt.plot([-2], [0], 'ro', markersize=8, label='Point (-2, 0)')
plt.plot([0], [1], 'ro', markersize=8, label='Point (0, 1)')
plt.plot([-1], [3], 'ko', markersize=8, label='$y_1$ = -1')
plt.text(-1, 3.3, '$X_1, \\rho=5$', ha='center', fontsize=10)
plt.plot([-2], [1], 'bo', markersize=8, label='$y_2$ = -1')
plt.text(-2, 1.3, '$X_2, \\rho=2$', ha='center', fontsize=10)
plt.plot([1], [2], 'go', markersize=8, label='$y_3$ = -1')
plt.text(1, 2.3, '$X_3, \\rho=1$', ha='center', fontsize=10)
plt.plot([2], [1], 'yo', markersize=8, label='$y_4$ = 1')
plt.text(2, 1.3, '$X_4, \\rho=2$', ha='center', fontsize=10)
plt.plot([-1], [-1], 'co', markersize=8, label='$y_5$ = 1')
plt.text(-1, -0.7, '$X_5, \\rho=3$', ha='center', fontsize=10)

plt.arrow(0, 0, 1, -2, head_width=0.2, head_length=0.3, 
          fc='red', ec='red', length_includes_head=True, 
          label='$w^* = [1, -2]^{\\top}$')

plt.xlabel('$x_1$', fontsize=12)
plt.ylabel('$x_2$', fontsize=12)

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')

plt.savefig('homework_2_plot.png', dpi=300)

plt.show()
```
